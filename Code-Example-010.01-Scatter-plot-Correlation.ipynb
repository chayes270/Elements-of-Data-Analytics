```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Correlation

Is there an association between the number of hours spent studying and the performance on the final exam? 


```python
students = pd.read_csv("./datasets/students.csv")
students.info()
```

```python
students
```

```python
hours = students["hours"]
score = students["score"]
```

```python
fig = plt.figure()
fig.scatter(hours, score, c="red", alpha=0.9, marker='*',label="Score")
fig.xlabel("Hours of Study")
fig.ylabel("Exam Score")
fig.legend(loc='upper left')
# plt.show()
```

```python
plt.savefig('Correlation-hoursOfStudy-Score.png', 
            dpi=300, bbox_inches='tight')
plt.show()
# plt.savefig('Correlation-hoursOfStudy-Score.png', dpi=300)
```
