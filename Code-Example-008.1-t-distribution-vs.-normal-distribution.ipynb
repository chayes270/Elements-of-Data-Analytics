```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t
from scipy.stats import norm
```

```python
df = 5
x = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)
```

```python
fig, ax = plt.subplots(1,1)

ax.plot(x, t.pdf(x, df),'red', lw=3, alpha=0.6, label='t-distribution')
ax.plot(x, norm.pdf(x),'green', lw=3, alpha=0.6, label='normal distribution')
ax.legend(loc='best', frameon=False)
plt.show()

```
