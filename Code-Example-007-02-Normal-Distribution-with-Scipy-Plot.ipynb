```python
import numpy as np
import matplotlib.pyplot as plt

# from scipy.stats import norm
from scipy import stats

import statistics
```

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-100, 100, 1)

# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)

# We use PDF function to sample from this normal distribution. 
n_array = stats.norm.pdf(x_axis, mean, sd)

print(n_array)
```

```python
plt.plot(x_axis,n_array )
plt.show()
```

# Another Example using Numpy Random Normal 

```python
# 2000 samples
# mean = 0 
# st = 5 
n_data = np.random.normal(0, 5, 2000)

# the histogram of the data
n, bins, patches = plt.hist(n_data, 20, facecolor='g', alpha=0.50)

plt.grid(True)
plt.show()
```
