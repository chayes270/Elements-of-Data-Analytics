```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Correlation

Is there an association between the number of hours spent studying and the performance on the final exam? 


```python
ceo_df = pd.read_csv("./datasets/CEO_salary.csv")
ceo_df.info()
```

```python
ceo_df
```

```python
age = ceo_df["age"]
salary = ceo_df["salary"]
```

```python
fig = plt.figure()
plt.scatter(age, salary, c="red", alpha=0.9, marker='*',label="salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc='upper left')
plt.show()
```

```python
fig.savefig('Correlation-CEO.png', 
            dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Or save it as PDF file
fig.savefig('Correlation-CEO.pdf')
plt.show()
```
